```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("Dataset/Training Dataset.csv")
```

```python
df = data.copy()
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['Credit_History'].hist(bins=20)
```

```python
df['Loan_Amount_Term'].hist(bins=20)
```

```python
df['Gender'].fillna(df["Gender"].mode()[0],inplace=True)
df['Married'].fillna(df['Married'].mode()[0],inplace=True)
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0],inplace=True)
df['Dependents'].fillna(df["Dependents"].mode()[0],inplace=True)
df['Credit_History'].fillna(df['Credit_History'].mode()[0],inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0],inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['TotalIncome'] = df['ApplicantIncome']+df['CoapplicantIncome']
```

```python
df.head()
```

## Visualization

```python
df['Dependents'].unique()
```

```python
df['Dependents'] = df['Dependents'].map({'0':0,'1':1,"2":2,"3+":3})
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in ['Gender', 'Married', 'Education', 'Self_Employed', 'Loan_Status']:
    df[column] = le.fit_transform(df[column])
```

```python
df.drop(['Loan_ID','Property_Area'],inplace=True,axis=1)
```

```python
df.head()
```

```python
x = df.drop('Loan_Status', axis=1)
y = df['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.33, random_state=42)
```

```python
df.info()
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier() 
random_forest.fit(X_train,y_train)
```

```python
from sklearn import metrics
RF_predict = random_forest.predict(X_test)
```

```python
print(f"Accuracy of Random Forest classifier is: {metrics.accuracy_score(RF_predict,y_test)}")
```

```python
RF_predict
```

```python
from sklearn.naive_bayes import GaussianNB
naive = GaussianNB()
naive.fit(X_train,y_train)
```

```python
NB_predict = naive.predict(X_test)
print(f"Naive Bayes Accuracy is : {metrics.accuracy_score(NB_predict,y_test)}")
```

```python
from sklearn.tree import DecisionTreeClassifier
DT_class = DecisionTreeClassifier()
DT_class.fit(X_train,y_train)
```

```python
DT_predict = DT_class.predict(X_test)
print(f"Decision Tree Accuracy is : {metrics.accuracy_score(DT_predict,y_test)}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
KN_class = KNeighborsClassifier()
KN_class.fit(X_train,y_train)
```

```python
KN_predict = KN_class.predict(X_test)
print(f"KNeighborsClassifier Accuracy is : {metrics.accuracy_score(KN_predict,y_test)}")
```
